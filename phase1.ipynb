```python
def prepare_text(raw_text):
    raw_text = raw_text.lower()
    raw_text = re.sub(r'\d+', '' , raw_text)
    raw_text = re.sub (r'[^\w\s]' ,'', raw_text)
    raw_text = raw_text.strip()
    raw_text = raw_text.split()
    stemmer =  nltk.PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    raw_text = [stemmer.stem(word) for word in raw_text]
    raw_text = [lemmatizer.lemmatize(word, wordnet.VERB) for word in raw_text]
    return raw_text
```

```python
def get_stop_words(raw_text):
    raw_text = Counter(raw_text)
    word_dictionary = dict(raw_text)
    sort_orders = dict(sorted(word_dictionary.items(), key=lambda x: x[1], reverse=True))
    return sort_orders
```

```python
def remove_stop_words(raw_text, freq):
    words_to_remove=[]
    for i,d in enumerate(freq):
        words_to_remove.append(d)
        if i == 5:
            break
    result = [i for i in raw_text if i not in words_to_remove]
    
    return result
```

```python
dataSet = pd.read_csv("english_dataset.csv")
data = list(dataSet["title"]+" "+dataSet["text"])
res = []
def get_text(limit):
    limit = limit
    for i,d in enumerate(data):
        res.append(prepare_text(d))
        if i == limit:
            break
    return res
    for i in data:
        processed_data = prepare_text(i)
        stopped = get_stop_words(processed_data)
        removed = remove_stop_words(processed_data, stopped)
    print(removed)

text = prepare_text(data[0])
stop = get_stop_words(text)
remove = remove_stop_words(text,stop)
```

```python
def get_positional_index(raw_text):
    posting_list = {}
    doc_number = 0
    for index, word in enumerate(raw_text):
        if word in posting_list:
            posting_list[word][0] = posting_list[word][0] + 1
            if doc_number in posting_list[word][1]:
                 posting_list[word][1][doc_number].append(index)        
            else:
                posting_list[word][1][doc_number] = [index]
        else:
            posting_list[word] = []
            posting_list[word].append(1)
            posting_list[word].append({})     
            posting_list[word][1][doc_number] = [index]
    return posting_list
```

```python
unary = lambda n: n * '0' + '1'
binary = lambda n, l = 1:("{0:0%db}" % l).format(n)
def gamma_encode(n):
    if(n==0): 
        return '0'
    m = int(log(n, 2))
    u = m
    b = n - 2 ** m
    return unary(m) + binary(b, m)

def gamma_decode(n):
    unary = n[:n.find('1')+1]
    N = len(unary)-1
    binary = n[n.find('1')+1:]
    x = '1'+binary
    decimal_number = int(x,2)
    return decimal_number
```

```python
def convert_to_gamma_code(positional_indices):
    for word, posting_list in positional_indices.items():
        count, positions = posting_list
        new_posting_list = [count]
        for sentence_index, word_positions in positions.items():
            new_sentence_index = gamma_encode(sentence_index)
            new_word_positions = [gamma_encode(word_position) for word_position in word_positions]
            new_posting_list.append({new_sentence_index: new_word_positions})
        positional_indices[word] = new_posting_list
    return positional_indices
```

```python
def gamma_to_dict(positional_indices):
    for word, posting_list in positional_indices.items():
        count, positions = posting_list
        new_posting_list = [count]
        for sentence_index, word_positions in positions.items():
            new_sentence_index = gamma_decode(sentence_index)
            new_word_positions = [gamma_decode(word_position) for word_position in word_positions]
            new_posting_list.append({new_sentence_index: new_word_positions})
        positional_indices[word] = new_posting_list
    return positional_indices
```

```python
def encode_number(number):
    bytes_list = []
    while True:
        bytes_list.insert(0, number % 128)
        if number < 128:
            break
        number = number // 128
    bytes_list[-1] += 128
    return pack('%dB' % len(bytes_list), *bytes_list)

def encode(numbers):
    bytes_list = []
    for number in numbers:
        bytes_list.append(encode_number(number))
    return b"".join(bytes_list)

def decode(bytestream):
    n = 0
    numbers = []
    bytestream = unpack('%dB' % len(bytestream), bytestream)
    for byte in bytestream:
        if byte < 128:
            n = 128 * n + byte
        else:
            n = 128 * n + (byte - 128)
            numbers.append(n)
            n = 0
    return numbers
```

```python
def convert_to_variable_bytes(positional):
    for word, posting_list in positional.items():
        count , positions = posting_list
        new_posting_list = [count]
        for sentence_index, word_positions in positions.items():
            new_sentence_index = encode_number(sentence_index)
            new_word_positions = encode(word_positions)
            new_posting_list.append({new_sentence_index: new_word_positions})
        positional[word] = new_posting_list
    return positional
```

```python
def variable_byte_to_dict(positional_indices):
    for word, posting_list in positional_indices.items():
        count, positions = posting_list
        new_posting_list = [count]
        for sentence_index, word_positions in positions.items():
            new_sentence_index = decode(sentence_index)
            new_word_positions = [decode(word_position) for word_position in word_positions]
            new_posting_list.append({new_sentence_index: new_word_positions})
        positional_indices[word] = new_posting_list
    return positional_indices
```

```python

def editDistance(str1, str2, m, n):

    if m == 0:
        return n

    if n == 0:
        return m

    if str1[m-1] == str2[n-1]:
        return editDistance(str1, str2, m-1, n-1)

    return 1 + min(editDistance(str1, str2, m, n-1),    # Insert
                   editDistance(str1, str2, m-1, n),    # Remove
                   editDistance(str1, str2, m-1, n-1)    # Replace
                   )
```

```python

def jaccard(listA, listB):
    numerator = len(list(set(listA) & set(listB)))
    denominator = len(set(listA)) + len(set(listB)) - numerator
    return  numerator / denominator
```
